```python
##Concatenation of Data Frames

import pandas as pd
india_weather = pd.DataFrame({\
    "city": ["mumbai" , "delhi", "Banglore"] ,\
     "temperature" : [32,45,30], \
    "humidity" : [80,60,78]\
  })
```

```python
us_weather = pd.DataFrame({ 
    "city" : ["new york", "chicago" , "orlando"],\
    "temperature" : [21,14,35],\
    "humidity" : [68,65,75], \
     
})
```

```python
us_weather
```

```python
df = pd.concat([india_weather, us_weather])
```

```python
df
```

```python
dfC = pd.concat([india_weather , us_weather ], ignore_index = True)
```

```python
dfC
```

```python
dfAxis = pd.concat([india_weather , us_weather ], axis= 1)
```

```python
dfAxis
```

```python
#Merge data frames

temperature_df = pd.DataFrame({
    "city" : ["mumbai", "delhi", "banglore" , "hyd"],\
    "temperature" : [32,45,30,40]
})
```

```python
temperature_df
```

```python
humidity_df = pd.DataFrame({
    "city" : ["delhi" , "mumbai", "banglore"],\
    "humidity" : [68,65,75]\
    
})
```

```python
humidity_df
```

```python
#merging two data frames without explicitly mentioning any index

df = pd.merge(temperature_df , humidity_df, on = 'city')
```

```python
df
```

```python
#Through outer join
df = pd.merge(temperature_df, humidity_df , on = 'city', how = 'outer')
```

```python
df
```
